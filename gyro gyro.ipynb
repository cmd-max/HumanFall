```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Asus\Desktop\datasets\Train.csv")
```

```python
df_df = pd.read_csv(r"C:\Users\Asus\Desktop\datasets\Test.csv")
```

```python

df.drop('Unnamed: 0', axis=1, inplace=True)

df_df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
X_train = df.drop(['fall','label'],axis=1)
y_train = df['fall']
X_test =  df_df.drop(['fall','label'],axis=1)
y_test =  df_df['fall']
```

```python
SEED = 42
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=3, 
                             max_depth=2,
                             random_state=SEED)
```

```python
# Fit RandomForestClassifier
rfc.fit(X_train, y_train)
# Predict the test set labels
y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import classification_report

classification_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(classification_report)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
features = X_train.columns.values # The name of each column
classes = ['0', '1'] # The name of each class
# You can also use low, medium and high risks in the same order instead
# classes = ['low risk', 'medium risk', 'high risk']

for estimator in rfc.estimators_:
    print(estimator)
    plt.figure(figsize=(12,6))
    tree.plot_tree(estimator,
                   feature_names=features,
                   class_names=classes,
                   fontsize=8, 
                   filled=True, 
                   rounded=True)
    plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d').set_title('gyro')

print(classification_report(y_test,y_pred))
```
